```python
import os
import json
from datetime import datetime
```

```python
path = "/Users/albert/Documents/Finances/data/raw/alternative_data/news/idx_channel/detailed/ticker/idxchannel_keyword_BUMI_20260124_160327_DETAILED_20260124_165926.json"
```

```python
# Baca file JSON
with open(path, "r", encoding="utf-8") as f:
    data = json.load(f)

# Proses data
processed_data = [
    {
        "title": item.get("title"),
        "date": item.get("date"),
        "full_content": item.get("full_content"),
        "summary": item.get("summary"),
    }
    for item in data
]

# ðŸ”¹ Sorting berdasarkan date, dari terlama ke terbaru
# Jika date formatnya ISO 8601 (contoh: "2025-01-02T00:00:00")
processed_data = sorted(
    processed_data,
    key=lambda x: datetime.fromisoformat(x["date"])
)

# Simpan hasil
output_path = f"/Users/albert/Documents/Finances/data/processed/filtered_emiten/bumi_filtered_2025_2026_cleaned_news.json"

with open(output_path, "w", encoding="utf-8") as f:
    json.dump(processed_data, f, ensure_ascii=False, indent=2)
```
