# Concat all data previous 90 days 

```python
dataframes = []
main_dir = 'data/'

for folder in os.listdir(main_dir):
    folder_path = os.path.join(main_dir, folder)

    if os.path.isdir(folder_path):
        csv_files = glob.glob(os.path.join(folder_path, "*.csv"))
        for file_path in csv_files:
            df = pd.read_csv(file_path)
            dataframes.append(df)

if dataframes:
    combined_chat_90d_20260127 = pd.concat(dataframes, ignore_index=True)
    display(combined_chat_90d_20260127)
else:
    print("Tidak ada file CSV yang ditemukan.")
```

```python
combined_chat_90d_20260127 = combined_chat_90d_20260127[['date', 'message']]
combined_chat_90d_20260127 = combined_chat_90d_20260127.dropna(subset=['message'])

print(combined_chat_90d_20260127.info())

print(combined_chat_90d_20260127['message'].value_counts())
```

```python
def clean_text(text):
    text = re.sub(r'[^\x00-\x7F]+', '', text)  # Menghapus karakter non-ASCII
    text = re.sub(r'[^A-Za-z0-9\s]', '', text)  # Menghapus karakter khusus
    return text

def has_stock_code(msg):
    if pd.isna(msg):
        return False
    words = re.findall(pattern, msg)
    # Cek apakah ada kata yang ada di stock_set
    return any(word in stock_set for word in words)
```

```python
combined_chat_90d_20260127['message'] = combined_chat_90d_20260127['message'].apply(clean_text)

combined_chat_90d_20260127['message'] = combined_chat_90d_20260127['message'].replace('NaN', np.nan)
combined_chat_90d_20260127['message'] = combined_chat_90d_20260127['message'].replace('', np.nan)

combined_chat_90d_20260127 = combined_chat_90d_20260127.dropna(subset=['message'])
combined_chat_90d_20260127 = combined_chat_90d_20260127.reset_index()

print(combined_chat_90d_20260127['message'].value_counts())
print(combined_chat_90d_20260127)
```
